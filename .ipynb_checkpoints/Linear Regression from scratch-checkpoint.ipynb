### Imports

```python
import math
import matplotlib.pyplot as plt
import numpy as np
import random

%matplotlib inline
```

#### Generate Data

```python
# Generate a random number in range (-2.0, 2.0)
def random_delta():
    return (random.random() - 0.5) * 10.0

xs = [x for x in range(0, 100)]
ys = [(2.0*x + 10.0) + random_delta() for x in xs]
```

```python
# Plot the points, and line of best fit.
plt.plot(xs, ys)

ys_best = [(2.0*x + 10.0) for x in xs]

#ys_best = [(0.57*x + 0.1487) for x in xs]
plt.plot(xs, ys_best)
plt.show()
```

```python
def forward_prop(w, b, x):
    y_pred = [(w*x[i] + b) for i in range(len(x))]
    return y_pred

def back_prop(x, y_pred, y):
    N = len(y)
    
    # (2 / N) * sum(x_i * (y_pred - y_i))
    sum1 = sum([(y_pred[i] - y[i]) * x[i] for i in range(N)])
    dw = (2.0 / N) * sum1
        
    # (2 / N) * sum(y_pred_i - y_i)
    sum2 = sum([(y_pred[i] - y[i]) for i in range(N)])
    db = (2.0 / N) * sum2
    return dw, db
```

```python
def get_loss(y_pred, y):
    loss = 0.0
    N = len(y)
    
    for i in range(N):
        diff = y_pred[i] - y[i]
        loss = loss + diff**2
    
    return loss / N
```

```python
EPOCHS = 100000
lr = 0.0002
N = len(xs)

w = random.random()
b = 0.0

for _ in range(EPOCHS):
    y_pred = forward_prop(w, b, xs)
    
    # update the weights and biases using gradient descent
    dw, db = back_prop(xs, y_pred, ys)
    w = w - lr*dw
    b = b - lr*db
    
    # Print the loss
    loss = get_loss(y_pred, ys)
    
    #print("w: {}, b: {}".format(w, b))
    #print("dw: {}, db: {}".format(dw, db))
    #print("loss: {}".format(loss))
```

```python
plt.plot(xs, ys)

ys_best = [(w*x + b) for x in xs]
plt.plot(xs, ys_best)
plt.show()

y_pred = forward_prop(w, b, xs)
print("loss: {}".format(get_loss(y_pred, ys)))
print("w: {}, b: {}".format(w, b))
```
